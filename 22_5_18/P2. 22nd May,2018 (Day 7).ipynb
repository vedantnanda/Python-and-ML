```python
#imports required for functions
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x = np.linspace(0,5,10) #10 sample points from 0 to 5
print(x) #will print the array
print(type(x))
y = x**2 #the equationy is x reaised to 2

figure() #after this the grpah is described
plot(x,y,'r') #plot x w.r.t y and 'r'-red 
xlabel('x')
ylabel('y')
title('title')
show() #mandatory without this won't show output
```

```python
#subplot
subplot(1,2,1) #1 row 2 columns 1st slot
plot(x,y,'r--') # x wrt y and r-- dashed line in red
subplot(1,2,2) #1 row 2 columns 2nd slot
plot(y,x,"g*-") #y wrt x *- style green line
```

Plot the equation: y=3*x +7 (plot(x,y...))<br>

put 4 plots in 1 graph

```python
x = np.linspace(5,50,150)
y = 3 * x + 7

figure()
subplot(2,2,1)
plot(y,x,'b--')
subplot(2,2,2)
plot(x,y,'m+')
show()

m = np.linspace(3,123,125)
n = 5 * m ** 2 + 17
figure()
subplot(2,2,3)
plot(m,n,':g')
subplot(2,2,4)
plot(n,m,'*-y')
show()
```
