## Counting airline and crashes

```python
#Number of times each unique data occurs 
new_data['Operator'].value_counts()
```

```python
grouped_data = new_data.groupby(['Operator'])

for name,group in grouped_data:
    print("Name = %s, Count = %d"%(name,len(group)))
```

```python
new_data['Location'].value_counts()
```

```python
grouped_data2 = new_data.groupby(['Operator','Location'])

for name,group in grouped_data2:
    if (len(group)>1):
        print("Name = %s, Count = %d"%(name,len(group)))
```

## Plot Histogram/Bar Chart of Airlines Name vs Number of crashes

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
new_data = data.dropna()
an = new_data['Operator'].value_counts()
an1 = dict(an.head(10))
fig,axes = plt.subplots(figsize=(15,7))
axes.bar(an1.keys(),an1.values(),width=0.8,color="red")
axes.set_xlabel('Airline Names')
axes.set_ylabel('Number of Crashes')
axes.set_title('Airlines Name vs Number of crashes')
axes.grid(True)
plt.xticks(rotation=90)
```

```python

import matplotlib
import matplotlib.pyplot as plt
import numpy

objects=new_data["Operator"].value_counts().keys()[0:10]
plt.bar(range(len(objects)),new_data["Operator"].value_counts().iloc[0:10],align="center")
plt.xticks(range(len(objects)),objects,rotation=90)
plt.title("Top 10 airline crashes")
plt.show()
```

## Plot Bar Chart of Airlines Name vs Total People Aboard all Flights

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
import pandas as pd
data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
new_data = data.dropna()
#new_data.groupby(['Operator'])['Aboard'].sum()
di = dict((new_data.groupby(['Operator'])['Aboard'].sum()[0:10]))

fig,axes = plt.subplots(figsize=(15,9))
axes.bar(di.keys(),di.values())
axes.set_xlabel('Airline Names')
axes.set_ylabel('Total people Aboard')
axes.set_title('Airlines Name vs Total people Aboard')
axes.grid(True)
plt.xticks(rotation=90)
```

## Plot a Scatter Plot of Aboard vs Fatalities

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
new_data = data.dropna()

fig,axes = plt.subplots(figsize=(15,7))
axes.scatter(new_data['Aboard'],new_data['Fatalities'])
axes.set_title("Aboard vs Fatalities")
axes.grid(True)
axes.set_xlabel('Aboard')
axes.set_ylabel('Fatalities')
```

```python
plt.scatter(new_data['Aboard'],new_data['Fatalities'])
plt.xlabel('Number of passangers Aboard')
plt.ylabel('Number of passangers died')
plt.title("Plot of Aboard vs Died")
plt.grid()
```
