## Association Analysis
- Simplest Form of analysis
- Given a set of transactions, find results that will predict 
  the occurance of an item based on the occurances of other items  in the transaction
<br><br>
#### Market Basket Analysis
<pre>

### Association Rule
- AN impllication expression of the form X-> Y<br>
  where X and Y are itemset.
- Example<br>
  {milk,diaper}-> beer

### Rule Evaluation Matrix
- Support(s)
- Confidence(c)

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import numpy as np
dataset = [["Bread", "Milk","Beer"],
          ["Bread","Diapers","Eggs"],
          ["Milk","Diapers","Beer","Cola"],
          ["Bread","Milk","Diapers","Beer"],
          ["Bread","Milk","Cola"]]
```

```python
te = TransactionEncoder()
te = te.fit(dataset) #extracts all the unique data from dataset
te_ary = te.transform(dataset) #converts array into hot coded matrix
print(te.columns_) 
print(te_ary)

df = pd.DataFrame(te_ary,columns  = te.columns_)
#print(df)

frequent_itemsets = apriori(df,min_support=0.3,use_colnames = True)
print(frequent_itemsets) #frequency of items


from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets,metric="support",min_threshold=0.4)

print(rules)
```

## Application

1. Take an input from user
2. Based on the user input check that item in antecedant and suggest its consequent based on confidence
3. If user doesn't select the suggested item, let user enter another item
4. Suggest user another item based on user's item list and suggested item should not be the user's list
5. Keep repeating the process till user stops

### How to proceed:
1. load dataset
2. Get rules from purchase history filter based on confidence levels
3. display items that can be bought by the user, if there are no items then stop purchase
4. Take input item from user
5. Suggest consequent based on antecedant presence in the filtered rules
6. Take another user input, and determine if it is in suggestions, go back to step 3
   stop if user wants to stop purchase
