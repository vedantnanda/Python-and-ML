```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x = np.linspace(0,5,10) #10 sample points from 0 to 5
print(x) #will print the array
print(type(x))
y = x**2 #the equationy is x reaised to 2

figure() #after this the grpah is described
plot(x,y,'r') #plot x w.r.t y and 'r'-red 
xlabel('x')
ylabel('y')
title('title')
show() #mandatory without this won't show output
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height
#place left bottom corner of axes at 0.1,0.1 coordinates ofcanvas
#will have width and height 0.8 times of full canvas

#fig = plt.figure()
#axes = fig.add_sample(111) # 1 row 1 column 
#231 - 2 rowa 3 columns 1st position
axes.plot(x,y,'r') 
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

#main figure
axes1.plot(x,y,'r') 
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

#insert
axes2.plot(y,x,'g') 
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
```

```python
fig, axes = plt.subplots(nrows=1,ncols = 2)
#print(plt.subplots(nrows=1,ncols = 2))
print(fig,"\n",axes)
print(type(axes))
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('y')
    ax.set_ylabel('x')
    ax.set_title('insert title')
```

## Setting colors, linewidths,linetypes

```python
#MATLAB style line color and style
ax.plot(x,x**2,'b.-')
ax.plot(x,x**3,'g--')

fig,ax = plt.subplots()

ax.plot(x,x+1,color="red",alpha=0.5) #half transparant red , alpha ranges 0 to 1
ax.plot(x,x+2,color="#1155dd") #RGB hex code for blue
ax.plot(x,x+3,color="#15cc15") #RGB hex code for green
```

```python
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(x,x+1,color="blue",linewidth=0.25)
ax.plot(x,x+2,color="blue",linewidth=0.50)
ax.plot(x,x+3,color="blue",linewidth=1.25)
ax.plot(x,x+4,color="blue",linewidth=2.25)

#possible linestype options '-' , '--' , '-.' , ':' , 'steps'
ax.plot(x,x+5,color="red",lw=2,linestyle="-")
ax.plot(x,x+6,color="red",lw=2,linestyle="-.")
ax.plot(x,x+7,color="red",lw=2,linestyle=":")

#custom dash
line, = ax.plot(x,x+8,color = 'black',lw = 1.50)
line.set_dashes([5,10,15,10]) #format: line length,space,length,space

#possible marker symbols: marker = '+','o','*','s',',','1'
ax.plot(x,x+9,color = 'green',lw = 2,ls="--",marker='+')
ax.plot(x,x+10,color = 'green',lw = 2,ls="--",marker='o')
ax.plot(x,x+11,color = 'green',lw = 2,ls="--",marker='s')
ax.plot(x,x+12,color = 'green',lw = 2,ls="--",marker='1')

#marker size and color
ax.plot(x,x+13,color = 'purple',lw = 1,ls="--",marker='o',markersize=2)
ax.plot(x,x+14,color = 'purple',lw = 1,ls="--",marker='o',markersize=4)
ax.plot(x,x+15,color = 'purple',lw = 1,ls="--",marker='o',markersize=8,markerfacecolor="red")
ax.plot(x,x+16,color = 'purple',lw = 1,ls="--",marker='s',markersize=8,markerfacecolor="yellow",
       markeredgewidth=2,markeredgecolor="blue")
```

## Axes Range

```python
fig,axes = plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].set_title("default axes range")

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title("custom axes range")
```

## Axis Grid

```python
fig,axes = plt.subplots(1,2,figsize=(10,5))

#default grid apperance
axes[0].plot(x,x**2,x,x**3,lw=2)
axes[0].grid(True)

#custom grid appreance
axes[1].plot(x,x**2,x,x**3,lw=2)
axes[1].grid(color='b',alpha=0.5,ls='dashed',lw=0.5)
```

## 2D plot styles

```python
print(np.random.randn(10))
```

```python
n = np.array([0,1,2,3,4,5])
x = np.linspace(0,5,10)
xx = np.linspace(-0.75,1.,100)

print(type(x))
print(len(xx))

fig,axes = plt.subplots(1,4,figsize=(15,3))

axes[0].scatter(xx,xx+0.75*np.random.randn(len(xx)),color="red")
axes[0].set_title("Scatter")

axes[1].step(n,n**2,lw=2)
axes[1].set_title("Step")

axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.8,color="black") #1 means occupy full align="edge"
axes[2].set_title("Bar")
axes[2].grid(True)

axes[3].fill_between(x,x**2,x**3,color="green",alpha=0.5)
axes[3].set_title("Fill between")
```

## Histogram

```python
n = np.array([1,1,2,2,2,2,3,4,5,5,6,6,7,8,8,9,9,10,10,10])
#1 - 2
#2 - 4
#3 - 1
#4 - 1
#5 - 2
#6 - 2
#7 - 1
#8 - 2
#9 - 2
#10 - 3

print(len(n))
print(min(n))
print(max(n))
print(n)
print(type(n))

fig,axes = plt.subplots(figsize = (12,4))

axes.hist(n)
axes.set_title("Histogram")
axes.set_xlim((min(n),max(n)))

```

```python
n = np.array([1,1,2,2,2,2,3,4,5,5,6,6,7,8,8,9,9,10,10,10])

print(len(n))
print(min(n))
print(max(n))
print(n)
print(type(n))

fig,axes = plt.subplots(1,2,figsize = (12,4))

axes[0].hist(n)
axes[0].set_title("Histogram")
axes[0].set_xlim((min(n),max(n)))

axes[1].hist(n,cumulative=True,bins=25) #scumulativve adding all previous so adds up last bin has sum
#bins is maximum number of rectangles to be used
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n),max(n)))
```
