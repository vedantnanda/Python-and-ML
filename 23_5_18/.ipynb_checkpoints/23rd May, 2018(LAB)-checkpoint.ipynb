### LA-27. Inverse of Matrix
Find inverse of a 2x2 matrix without using inverse function and print it. Take matrix values from user

```python
import numpy as np
a = []
b = []
try:
    a = list(map(float,input("Enter first row(space separated): ").split()))
    b = list(map(float,input("Enter second row(space seperated): ").split()))
    if len(a)==2 and len(b)==2:
        m = np.array([a,b])
        print(m)
        dino = (m[0][0]*m[1][1])-(m[0][1]*m[1][0])
        if dino != 0:
            d = 1/dino
            n = np.array([ [m[1][1],-1*(m[0][1])] , [-1*(m[1][0]),m[0][0]] ])
            print(d*n)
        else:
            print("Inverse doesnt exist")
    else:
        print("Invalid input")
except ValueError:
            print("Only Numbers allowed")
```

### LA-29.Determinant Luck Game
Team A and B both get 2 chances each to generate determinant of a random 2x2 matrix<br>
The Team with minimum determinant score will win.<br>
Print the score by which they lead

```python
import numpy as np
sa = 0
sb = 0
print("\t Determinant values","\n")
for i in range(10):
    a = np.random.randint(11, size=(2, 2))
    b = np.random.randint(11, size=(2, 2))
    d1 = (a[0][0]*a[1][1])-(a[0][1]*a[1][0])
    d2 = (b[0][0]*b[1][1])-(b[0][1]*b[1][0])
    print("chance {0}\t".format(i+1))
    sa = sa+d1
    sb = sb+d2
    print("   Team A score {0}     Team B score {1}\n\n".format(sa,sb))
if sa>sb:
    print("Team A wins by {0} points".format(sa-sb))
elif sa==sb:
    print("Match draw")
else:
    print("Team B wins by {0} points".format(sb-sa))
```

### LA-30. Plotting combined signal
Plot two different signals(for ex. Sine and cosine wave)<br>
and the combination of those signals combinations of sine, cos, tan

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

t = arange(0.0, 2, 0.01)
t1 = np.linspace(-2,2,100)
figure()
subplot(3,3,1)
plot(t1, sin(2*pi*t1),'r')
title("sin")
subplot(3,3,2)
plot(t1, cos(2*pi*t1),'g')
title("cos")
subplot(3,3,3)
plot(t1, sin(2*pi*t1)+cos(2*pi*t1),'m')
title("sin+cos")

subplot(3,3,4)
plot(t1, sin(2*pi*t1),'r')
title("sin")
subplot(3,3,5)
plot(t1, tan(t1))
title("tan")
subplot(3,3,6)
plot(t1, sin(2*pi*t1)+tan(t1),'m')
title("sin+tan")

subplot(3,3,7)
plot(t1, cos(2*pi*t1),'g')
title("cos")
subplot(3,3,8)
plot(t1, tan(t1))
title("tan")
subplot(3,3,9)
plot(t1, cos(2*pi*t1)+tan(t1),'r')
title("cos+tan")

show()
```

### LA31. System Matrix
The rating of acertain system is determinant * sum of all elements and is<br>
dependant on random values of elements. Write a code to print<br>
the value of rating for 5 continuous readings.<br>
Save the reading as dictionary <b>{determinant:rating}</b>

```python
import numpy as np
di = {}
for i in range(5):
    a = np.random.randint(11, size=(3, 3))
    de = (a[0][0]*((a[1][1]*a[2][2])-(a[1][2]*a[2][1]))) - (a[0][1]*((a[1][0]*a[2][2])-(a[2][0]*a[1][2]))) + (a[0][2]*((a[1][0]*a[2][1])-(a[1][1]*a[2][0])))
    r = de * a.sum()
    di[de] = r
print(di)
```

### LA-33 Matrix orthogonality check
Prove that the matrix is orthogonal if the inverse of a 2x2 matrix equal to its transpose.<br>
Take the matrix values from the user

```python
import numpy as np
a = []
b = []
try:
    a = list(map(float,input("Enter first row(space separated): ").split()))
    b = list(map(float,input("Enter second row(space seperated): ").split()))
    if len(a)==2 and len(b)==2:
        m = np.array([a,b])
        print(m)
        dino = (m[0][0]*m[1][1])-(m[0][1]*m[1][0])
        if dino != 0:
            d = 1/dino
            n = np.array([ [m[1][1],-1*(m[0][1])] , [-1*(m[1][0]),m[0][0]] ])
            print(d*n)
            inver = np.array(d*n)
            print(inver.T)
            if(np.array_equal(inver,inver.T)):
                print("Orthogonal")
            else:
                print("Not Orthogonal")
        else:
            print("Inverse doesnt exist")
    else:
        print("Invalid input")
except ValueError:
            print("Only Numbers allowed")
```

### LA-34. Chess Pattern
Create chess pattern of 1s and 0s using numpy

```python
import numpy as np
def chess(n):
    return np.array([[(i+j)%2 for i in range(n)] for j in range(n)])

print(chess(8))
```
